# Controlled vocabulary
The Microcompartment Database uses a controlled vocabulary to ensure consistency and accuracy in data entry and retrieval. This vocabulary includes standardised terms and definitions for various aspects of microcompartments, such as types, functions, and associated proteins. By adhering to this controlled vocabulary, users can effectively search and analyse the database, facilitating better communication and understanding within the scientific community.

# Restricted Vocabulary done via Enums - enums.py file

* DatabaseType: Types of databases available.
* StructProtType: Types of structural proteins.
* ModificationType: Types of protein modifications.
* ComplexSource: Sources of protein complexes.

The definitions of these enums can be found in the `enums.py` file.

**Vocabulary Definitions:**

* **DatabaseType**: This enum classifies the types of databases available in the Microcompartment Database. The types of databases include:

    * ***Sequence***: Databases that store protein or nucleotide sequences.
    * ***Structure***: Databases that store structure related information of proteins or other macromolecules.
    * ***Function***: Databases that store functional information about proteins or other biological entities.
    * ***Taxonomy***: Databases that store taxonomic information about the protein or other biological entities organism origin.

* **StructProtType**: This enum classifies the protein type, focusing on structural proteins. Hexamers, Trimers and Pentamers are the main structural proteins forming the microcompartment shell.The classification is:

    * ***Hexamer***: Hexamers and hexameric-like proteins (i.e. BMC-H, BMC-Hp).
    * ***Trimer***: Trimers and trimeric-like proteins (i.e. BMC-Ts).
    * ***Pentamer***: Pentamers and pentameric-like proteins (i.e. BMC-Px).
    * ***Other*** : Non-structural proteins, including core enzyme proteins, support proteins, and any other proteins not directly involved in forming the microcompartment shell.


* **ModificationType**: This enum classifies the types of modifications that can be made to proteins. The types of modifications include:

    * ***Truncated***: proteins that have been shortened by removing certain segments.
    * ***Extended***: proteins that have been lengthened by adding additional segments.
    * ***Fusion***: proteins that are a combination of two or more structural protein types.
    * ***Synthetic***: proteins that have been artificially designed or synthesized.
    * ***Mutated***: proteins that have undergone specific point mutations or internal changes.
    * ***Domesticated***: proteins that have been adapted for specific functions or vectors.

* **ComplexSource**: This enum classifies the sources of protein complexes. The sources include:
    * ***Native***: Protein complexes that are found naturally in organisms without any modifications.
    * ***Engineered***: Protein complexes that have been modified or designed for specific purposes.
    * ***Predicted***: Protein complexes that have been computationally predicted based on sequence or structural data.
    * ***Theoretical***: Protein complexes that are proposed based on theoretical models or hypotheses, which may not have experimental evidence yet.
