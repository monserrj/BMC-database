
# microcomp_db.py

Main entry point for the BMC Database project. Handles:

* Database creation and schema initialization
* Loading approved external database definitions from CSV
* Reading protein data from CSV input
* Linking protein data to database with cross-references
* Password-protected addition of unknown external databases

Uses Click for CLI, SQLAlchemy for ORM, and custom CSV processing.

## Running the script

```bash
python microcomp_db.py --dbpath my_database.db --csvpath data/proteins.csv --dbinfo data/databases.csv --force --verbose
```

## Available Options

### --dbpath PATH
Path where the SQLite database will be created.
* **Default**: `bmc.db` (in current folder)
* **Type**: File path (created if doesn't exist)

### --csvpath PATH
Path to the CSV file with protein data to be imported.
* **Default**: `../data/raw/prot_info/prot_data_minimal_correct_UP.csv`
* **Type**: File path (must exist)

### --dbinfo PATH
Path to the CSV file containing approved external database definitions.
* **Default**: `None` (no approved databases pre-loaded)
* **Type**: File path (must exist if specified)
* **Behavior**: Databases from this CSV are loaded without password prompts. Unknown databases require authentication.

### --force / --no-force
Controls database file overwriting behavior.
* **Default**: `--no-force`
* **--force**: Overwrites existing database file
* **--no-force**: Exits if database file already exists

### --verbose / --no-verbose
Controls output detail level.
* **Default**: `--no-verbose`
* **--verbose**: Shows detailed logging and CSV preview
* **--no-verbose**: Shows only essential progress messages

## Workflow

1. **Database Setup**: Create or overwrite SQLite database file based on --force option
2. **Schema Creation**: Initialize all database tables (Protein, Xdatabase, Xref, etc.)
3. **Approved Database Loading** (if --dbinfo specified): Load trusted external databases from CSV without passwords
4. **CSV Data Reading**: Parse protein data from specified CSV file
5. **Data Processing**: For each protein entry:
   - Add protein sequence and metadata to database
   - Process external database cross-references
   - For approved databases: Link directly
   - For unknown databases: Prompt for password and additional information (URL, type)
6. **Transaction Management**: Commit changes or rollback on errors

## Usage Examples

### Basic usage (with defaults)
```bash
python microcomp_db.py
```

### Full featured import with approved databases
```bash
python microcomp_db.py \
  --dbpath bmc_database.db \
  --csvpath ../data/raw/prot_info/prot_data_complete_UP.csv \
  --dbinfo ../data/raw/prot_info/db_info_ext.csv \
  --force \
  --verbose
```

### Import without approved databases (prompts for unknowns)
```bash
python microcomp_db.py \
  --dbpath new_database.db \
  --csvpath ../data/raw/prot_info/prot_data_minimal_correct_UP.csv \
  --force
```

### Get help
```bash
python microcomp_db.py --help
```

## Security Features

- **Approved Databases**: Databases listed in --dbinfo CSV are trusted and added without authentication
- **Unknown Databases**: Require password authentication (`DB_ADD_PASSWORD` environment variable) and complete information (URL, type)
- **Transaction Safety**: All operations use database transactions with automatic rollback on errors
