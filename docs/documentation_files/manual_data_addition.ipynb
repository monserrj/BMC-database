# Addition of Data Manually to the Database

The database supports manual data addition through helper functions and CSV-based imports.

## Core Modules

**Module: db.py**

Core database functions for adding data with deduplication and relationship management.

*Functions:*
* `add_protein(session, protseq, struct, canonical)` - Add proteins with structure type validation
* `add_xdatabase(session, xname, xurl=None, xtype=None, require_password=True)` - Add external databases with optional password protection
* `add_xref(session, xdb, protein, xrefacc, cds=None)` - Link proteins and CDS to external database references

---

**Module: file_and_data.py**

CSV-based data import and batch processing.

*Functions:*
* `load_databases_from_csv(session, dbinfo_path)` - Load approved external databases from CSV
* `link_db_csv(mydata, session, dbinfo_path=None)` - Process protein data from CSV and link to database with CDS and external references

---

**Module: readfile.py**

File reading utilities for CSV/Excel data.

*Functions:* 
* `read_file(filepath, verbose=False)` - Read and parse CSV files for data import with CLI

---

**Entry Point: microcomp_db.py**

Command-line interface for orchestrating the complete workflow:
```bash
python microcomp_db.py --dbpath bmc.db --csvpath data/proteins.csv --dbinfo databases.csv --force --verbose
```

*Workflow:*
1. Creates or overwrites database schema
2. Loads approved databases from CSV (if provided)
3. Reads protein data with CDS and external references
4. Links all data to database with automatic deduplication
